```python
# Importando bibliotecas
import pandas as pd
import seaborn as sns
import numpy as np
import chart_studio as py
import cufflinks as cf
 
import plotly.graph_objects as go
import plotly.express as px

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
from sklearn.metrics import r2_score

import matplotlib.pyplot as plt
```

```python
# Importando base de dados
base_casa = pd.read_csv("USA_Housing.csv", sep=",")
base_casa.head()
```

```python
# Informações da base
base_casa.info()
```

```python
# Descrevendo a base
base_casa.describe()
```

```python
# Descrevendo a Avg. Area Income de maneira visual
fig = px.box(base_casa, x="Avg. Area Income")
fig.update_traces(line_color="blue")
```

```python
# Descrevendo o Price de maneira visul
fig = px.box(base_casa, x="Price")
fig.update_traces(line_color="green")
```

```python
# Retirando espaços das colunas e substituindo por underscore
base_casa = base_casa.rename(columns={"Avg. Area Income": "Avg_Area_Income",
                                      "Avg. Area House Age": "Avg_Area House_Age",
                                      "Avg. Area Number of Rooms": "Avg_Area_Number_of_Rooms",
                                      "Avg. Area Number of Bedrooms": "Avg_Area_Number_of_Bedrooms",
                                      "Area Population": "Area_Population",
                                      }
                                      )
base_casa.head()
```

```python
# Removendo coluna Address do dataset
base_casa.drop(["Address"], axis=1, inplace=True)
base_casa.head()
```

```python
# Verificando correlações
sns.pairplot(base_casa)
```

```python
base_casa.columns
```

```python
# Comparação de correção somente com o preço
sns.pairplot(base_casa, x_vars=['Avg_Area_Income', 'Avg_Area House_Age', 
                                'Avg_Area_Number_of_Rooms',
                                'Avg_Area_Number_of_Bedrooms', 
                                'Area_Population'], y_vars="Price")
```

```python
# Verificando em forma de tabela, a correção das colunas
base_casa.corr()
```

```python
# Representação de heat map
sns.heatmap(base_casa.corr(), annot=True)
```

```python
# Distribuição dos preços
sns.histplot(base_casa["Price"]);
```

```python
# Separação de variáveis - Explicativas e base Y
x= base_casa[['Avg_Area_Income', 'Avg_Area House_Age', 'Avg_Area_Number_of_Rooms',
       'Avg_Area_Number_of_Bedrooms', 'Area_Population']]

y = base_casa[["Price"]]
```

```python
# Criando bases de teste e treino

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, 
                                                     test_size=0.3, 
                                                     random_state=42)
```

```python
# Imprimindo tamanhos das bases 
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Treinando modelo
lm.fit(x_train, y_train)
```

```python
# Predição na base de teste
y_pred = lm.predict(x_test)
```

```python
# Predição de acerto
r = r2_score(y_test, y_pred)

print("r_quadrado:", r)
```

```python
# Exbibição gráfica da predição
c = [i for i in range(1, 1501, 1)]
fig = plt.figure(figsize=(12, 8))
plt.plot(c, y_test, color="blue")
plt.plot(c, y_pred, color="red") # Cor da predição
plt.xlabel("index")
plt.ylabel("Preço Do Imóvel");
```
